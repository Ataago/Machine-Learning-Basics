# K means

```python
import pandas as pd

df = pd.read_csv('data/ch1ex1.csv')
df.head()
```

```python
from sklearn import cluster

mdl = cluster.KMeans(n_clusters=3)
mdl.fit(df)
labels = mdl.predict(df)
```

```python
import matplotlib.pyplot as plt

centroid = mdl.cluster_centers_

plt.scatter(df['0'], df['1'], c=labels)
plt.scatter(centroid[:,0], centroid[:, 1], c='r', marker='X', s=200)
```
