# Demonstration of Matplotlib Module

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn-whitegrid')
plt.axes()
```

## Plotting Sine Curves

```python
import numpy as np
x = np.linspace(0, 10, 50)

plt.plot(x, np.sin(x), label='Sin wave', c='r')
plt.xlim(0, np.pi * 2)
plt.ylim(-1.5, 1.5)

plt.legend()
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Wave');
```

```python

plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x), color=(1.0,0.2,0.3))   # RGB
plt.plot(x, np.sin(x - 1), color='#FFDD00') # HEX code
plt.plot(x, np.sin(x - 2), color='chartreuse') # HTML color names
```

## Plotting Lines

```python
linestyles = ['solid', 'dashed', 'dashdot', 'dotted', '-', '--', '-.', ':']
i = 0
for linestyle in linestyles:
    plt.plot(x, x + i, linestyle=linestyle)
    i += 1
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--r') # dashed cyan
```

## Scatter Plots

```python
plt.plot(x, np.sin(x), 'o')
plt.plot(x + 1, np.sin(x), '-o')
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, c='r', s=1000*y, alpha=0.4)
```

## Plotting Histogram

```python
data = np.random.normal(0, 10, 1000)  # (mean = 0, std = 10, n = 1000)
plt.hist(data);
```
