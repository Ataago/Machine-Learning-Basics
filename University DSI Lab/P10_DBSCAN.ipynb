# DBSCAN

```python
import numpy as np

from dataviz import generate_clusters, plot_clusters, plot_data
from dbscan import DBSCAN
```

```python
clusters = generate_clusters(4, 20, 7, (0, 100), (0, 100), seed = 1)
plot_data(clusters)
```

```python
dbscan = DBSCAN(eps=7, min_samples=5)
clusters = np.array(clusters)
dbscan.fit(clusters)
plot_clusters(clusters, labels=dbscan.labels_, components=dbscan.components_)
```
